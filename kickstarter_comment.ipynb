```python
def __comments(pro_url):
    driver.get(pro_url + "/comments")
#     driver.get("https://www.kickstarter.com/projects/crookedkings/orbiter-20-the-360-playing-card-display" + "/comments")
    driver.implicitly_wait(5)
    comments = ""
    try:

        while driver.find_element_by_css_selector("#react-project-comments > div > button"): 
            driver.find_element_by_css_selector("#react-project-comments > div > button").click()
            driver.implicitly_wait(5)
        driver.implicitly_wait(5)
        
    except:
        driver.implicitly_wait(5)
    try:
        mb2 = driver.find_element_by_css_selector("#react-project-comments > ul") 
        comments = mb2.text 
        comments = comments.split("\n")
        while "這個人已取消其認繳。顯示此留言。" in comments:
            comments.remove("這個人已取消其認繳。顯示此留言。")
    except:
        driver.implicitly_wait(5)
    return comments


```
